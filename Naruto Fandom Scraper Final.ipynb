```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
res = requests.get('https://naruto.fandom.com/wiki/Category:Characters')
res2 = requests.get('https://naruto.fandom.com/wiki/Category:Characters?from=Fugai%0AFugai')
res3 = requests.get('https://naruto.fandom.com/wiki/Category:Characters?from=H%C5%8Dki+%28Watari%29%0AH%C5%8Dki+%28Watari%29')
res4 = requests.get('https://naruto.fandom.com/wiki/Category:Characters?from=Kohari+Umino%0AKohari+Umino')
res5 = requests.get('https://naruto.fandom.com/wiki/Category:Characters?from=Nang%C5%8D%0ANang%C5%8D')
res6 = requests.get('https://naruto.fandom.com/wiki/Category:Characters?from=Shishio%0AShishio')
res7 = requests.get('https://naruto.fandom.com/wiki/Category:Characters?from=Uroko+Kurama%0AUroko+Kurama')

links = [res,res2,res3,res4,res5,res6,res7]
page = list()
for link in links:
    page.append(BeautifulSoup(link.text, 'html.parser'))
    print('link parsed!')
```

```python
atags = list()
imgtags= list()
main = list()
i=0
for html in page:
    atags=html.select('.category-page__member-link')
    imgtags = html.select('.category-page__member-thumbnail')
    for data in atags:
        #data.get('href')#get href of a tag
        #characterNames.append(data.getText())#get text inside the a tag
        for img in imgtags:
            if img.get('alt') == data.getText():
                #print(i,img.get('data-src'))
                # get the associated image of the name
                
                image = img.get('data-src')
                splitedImg=image.split('/revision')
                splitedImg.pop()
                #splitedImg[0]+='png'
                #print(splitedImg)
                
                main.append({'id':i,'link':data.get('href'),'name':data.getText(),'imageUrl':splitedImg[0]})
                i+=1
                break

        

pd.DataFrame(main)
```

```python
for i,link in enumerate(main):
    print(i,link)
```

```python
with open('characters.json','w') as f:
    f.write(json.dumps(main))
    print('success!!!!!')
    f.close()
```
